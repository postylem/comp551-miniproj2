```python
import pandas as pd
import numpy as np
from numpy import random

%matplotlib notebook
import matplotlib
import matplotlib.pyplot as plt


timing_df = pd.read_csv('accuracies_runtimes.csv')
clf_names = timing_df['Model']
indices = np.arange(len(timing_df))
score = timing_df['Accuracy']
training_time = timing_df['Fit_runtime']
test_time = timing_df['Predict_runtime']
```

```python
# a time plot
plt.figure(figsize=(7, 3))
plt.title("Train/test time")
# plt.barh(indices, score, .2, label="accuracy", color='navy')
plt.barh(indices, training_time, .2, label="training time", color='c')
plt.barh(indices + .3, test_time, .2, label="test time", color='darkorange')
plt.yticks(())
plt.legend(loc='best')
plt.subplots_adjust(left=.35)
# plt.subplots_adjust(top=.95)
# plt.subplots_adjust(bottom=.05)

for i, c in zip(indices, clf_names):
    plt.text(-25, i, c)
```
